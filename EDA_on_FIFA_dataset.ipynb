#### An Analysis of Manchester United

```python
manutd = fifa[fifa['club']=="Manchester United"]

manutd.head()
```

```python
manutd['nationality'].value_counts()
```

English players in Manchester United

```python
mutd_eng = manutd[manutd['nationality']=='England']
```

Foreign players in Manchester United

```python
mutd_forgn = manutd[manutd['nationality']!='England']
```

##### Analyzing salaries of foreign players and domestic players in Manchester United

```python
# Creating a plot
plt.figure(figsize=(10, 6))

# Plotting 'wage_eur' for mutd_eng 

plt.scatter(range(len(mutd_eng)), mutd_eng['wage_eur'], color='crimson', label='Domestic Players')
for i, short_name in enumerate(mutd_eng['short_name']):
    plt.annotate(short_name, (i, mutd_eng['wage_eur'].iloc[i]), fontsize=8, color='crimson')

# Plot 'wage_eur' for mutd_forgn 

plt.scatter(range(len(mutd_forgn)), mutd_forgn['wage_eur'], color='navy', label='Foreign Players')
for i, short_name in enumerate(mutd_forgn['short_name']):
    plt.annotate(short_name, (i, mutd_forgn['wage_eur'].iloc[i]), fontsize=8, color='navy')
    
# Customize the plot

plt.title("Wage Variation between Foreign and Domestic Players in Manchester United")
plt.xlabel("Player")
plt.ylabel("Wage (in Euros)")
plt.legend()



plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
mutd_eng.sort_values(by=['wage_eur'],ascending=False)[0:5]
```

```python
# Top 5 English wage earners of Man Utd

mutd_eng.sort_values(by=['wage_eur'], ascending=False).iloc[0:5][['long_name', 'wage_eur']]
```

```python
# Top 5 foreign players according to wage earnings

mutd_forgn.sort_values(by=['wage_eur'],ascending=False).iloc[0:5][['short_name','wage_eur']]
```

##### Best shot amongst English players at Man Utd

```python
top_10_players= mutd_eng.sort_values(by=['shooting'],ascending=False)[0:10][['long_name','shooting','wage_eur']]
```

```python
# Extracting the desired columns

player_names = top_10_players['long_name']
shooting_values = top_10_players['shooting']
wage_values = top_10_players['wage_eur']

# Creating a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(shooting_values, wage_values, marker='o', color='crimson')

# Adding labels for each point (player name)
for i, name in enumerate(player_names):
    plt.annotate(name, (shooting_values.iloc[i], wage_values.iloc[i]), fontsize=10)


plt.title("Relationship between Shooting and Wage for English Players at Manchester United")
plt.xlabel("Shooting")
plt.ylabel("Wage (in Euros)")

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top_10_players=mutd_eng.sort_values(by=['defending'],ascending=False)[0:5][['long_name','defending','wage_eur']]
```

```python
player_names = top_10_players['long_name']
shooting_values = top_10_players['defending']
wage_values = top_10_players['wage_eur']

# Creating a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(shooting_values, wage_values, marker='o', color='crimson')

# Adding labels for each point (player name)
for i, name in enumerate(player_names):
    plt.annotate(name, (shooting_values.iloc[i], wage_values.iloc[i]), fontsize=10)


plt.title("Relationship between Defending and Wage for English Players at Manchester United")
plt.xlabel("Shooting")
plt.ylabel("Wage (in Euros)")

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
